Las librerias siguientes son usadas para varias cosas. La primera es para calculos numericos, la segunda es para calculo simbolico y la ultima es para imprimir


```python
import numpy as np
import sympy as sy
import matplotlib.pyplot as plt
```

Las siguientes son funciones a usar para efectos practicos. La primera es una exponencial con un coseno, la segunda es una expresion de la funcion primera, pero en serie de Taylor

```python
x, y, z = sy.symbols('x y z') 

x=sy.symbols('x')

f=sy.exp(x)-3*sy.cos(x) #una funcion
g=sy.series(f,x,0,3) #serie de taylor alrededor de cero hasta orden 3 de la funcion anterior


#imprime las funciones como un tipo de formula

print(f)
print(g)
```

```python
order=g.getO() #imprime la funcion con "O"
gx=g.removeO() #funcion para poder obtener "O"

print(g) #imprime la funcion total

print(order) #imprime la funcion O exponente residuo

print(gx) #imprime la funcion sin el error
```

```python
#func=sy.lambdify(x, expr) esta es una notacion 
#con esto se hacen cosas que debo entender

ffunc=sy.lambdify(x,f)
gfunc=sy.lambdify(x,g.removeO())


#con esto evaluo la funcion 
ffunc(3)
gfunc(3)


#con esto imprimo varias veces una evaluacion.
print(ffunc(3)) 
print(gfunc(3))
```

```python
X=np.linspace(-3,3)

Yf=ffunc(X)
plt.plot(X,Yf, label="f(x)", color="red")


Yg=gfunc(X)
plt.plot(X,Yg, label="g(x)", color="violet")

plt.title("Funciones y error")
plt.legend(['f(x)','g(x)'])
plt.grid(True)
plt.show()


#plt.xlabel("xd")
#plt.ylabel("f(x),g(x)")
```

```python
Y_err=np.abs(Yf-Yg)
Y_ord=X**3
plt.plot(X, Y_err, color="blue")
plt.plot(X, Y_ord, color="violet")
plt.legend(['Error', 'Orden'])
plt.grid(True)
plt.show()#imprimir graf
```
