Se comienza de manera usual, usando estas tres librerias

```python
import numpy as np
import sympy as sy
import matplotlib.pyplot as plt
```

Luego, al haber declarado las librerias, se leige una variable y despues una funcion cualquiera. En este caso es una muy conocida, y esta misma es expresada como una serie de Taylor alrededor de cero hasta un orden 5 de O mayuscula.

Despues de esto, se imprimen de forma simbolicxa las funciones. La funcion "f" es como tal la funcion declarada en un inicio mientras que la funcion "g" expresa a la funcion "f" como una serie de Taylor, donde O(x^5) indica el error de grado 5.

```python
x=sy.symbols('x')

f=(sy.sin(x))/x #una funcion principal e inicial

g=sy.series(f,x,0,5) #serie de taylor alrededor de cero hasta orden 5 de la funcion anterior


#imprime las funciones como un tipo de formula

print(f)
print(g)
```

Para poder declarar el apartado de error, se siguen las sentencias a continuacion, donde al imprimirlas de manera individual, estas expresan:

1. La funcion g original.
2. La expresion de error
3. La expresion sin el calculo de error.

```python
order=g.getO()
gx=g.removeO()

print(g)

print(order)

print(gx)
```

Luego, se puede hacer uso de las siguientes sentencias, las cuales muestran de forma numerica los valores de las funciones "f" y "g" evaluadas en el valor 0.1. En la primera se evalua la funcion total mientras que en la segunda se evalua pero sin la O mayuscula.

```python
ffunc=sy.lambdify(x,f)
gfunc=sy.lambdify(x,g.removeO())


print(ffunc(0.1))
print(gfunc(0.1))
```

Finalmente, se muestran las graficas de las funciones descruitas a lo largo de este proceso.

La primera a continuacion son las graficas sobrepuestas de la funcion "f"original sobre la funcion "g" en forma de serie de Taylor. Los valores de las abcisas depende directamente de la cantidad de curva a visualizar.

```python
X=np.linspace(-3,3)

Yf=ffunc(X)
plt.plot(X,Yf, label="f(x)", color="red")


Yg=gfunc(X)
plt.plot(X,Yg, label="g(x)", color="violet")

plt.title("Funciones y error")
plt.legend(['f(x)','g(x)'])
plt.grid(True)
plt.show()
```

Y esta ultimas grafica muestra la comparacion entre los errores de la propia computadora contra la expresion en serie.

```python
Y_err=np.abs(Yf-Yg)
Y_ord=X**3
plt.plot(X, Y_err, color="blue")
plt.plot(X, Y_ord, color="violet")
plt.legend(['Error', 'Orden'])
plt.grid(True)
plt.show()#imprimir graf
```
